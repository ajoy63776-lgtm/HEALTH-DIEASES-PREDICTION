```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
df =pd.read_csv('health_data.csv')

label_encoders = {}
for col in ['Gender', 'Smoking', 'FamilyHistory', 'Disease']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
X = df.drop('Disease', axis=1)
y = df['Disease']
model = RandomForestClassifier()
model.fit(X, y)
# [Age, Gender, BMI, BP, Sugar, Cholesterol, Smoking, FamilyHistory]
user_input_raw =  [60, 'Male', 30.1, 150, 200, 240, 'Yes', 'Yes']
user_input_encoded = [
    user_input_raw[0],
    label_encoders['Gender'].transform([user_input_raw[1]])[0],
    user_input_raw[2],
    user_input_raw[3],
    user_input_raw[4],
    user_input_raw[5],
    label_encoders['Smoking'].transform([user_input_raw[6]])[0],
    label_encoders['FamilyHistory'].transform([user_input_raw[7]])[0]
]
prediction = model.predict([user_input_encoded])
predicted_label = label_encoders['Disease'].inverse_transform(prediction)
print("\n Predicted Disease:", predicted_label[0])
```

```python
from flask import Flask, request, render_template_string
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
app = Flask(__name__)
df = pd.read_csv("/content/health_data.csv")
label_encoders = {}
for col in ['Gender', 'Smoking', 'FamilyHistory', 'Disease']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
X = df.drop('Disease', axis=1)
y = df['Disease']
model = RandomForestClassifier(random_state=42)
model.fit(X, y)
form_template = """
<!DOCTYPE html>
<html>
<head>
    <title>Health Prediction</title>
    <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0/dist/css/bootstrap.min.css" rel="stylesheet">
</head>
<body class="bg-light">
<div class="container mt-5">
    <div class="card shadow-lg p-4">
        <h2 class="text-center text-primary">Health Prediction</h2>
        <form action="/predict" method="post">
            <div class="mb-3">
                <label>Age</label>
                <input type="number" class="form-control" name="Age" required>
            </div>
            <div class="mb-3">
                <label>Gender</label>
                <select class="form-control" name="Gender" required>
                    <option>Male</option>
                    <option>Female</option>
                </select>
            </div>
            <div class="mb-3">
                <label>BMI</label>
                <input type="number" step="0.1" class="form-control" name="BMI" required>
            </div>
            <div class="mb-3">
                <label>BP</label>
                <input type="number" class="form-control" name="BP" required>
            </div>
            <div class="mb-3">
                <label>Sugar</label>
                <input type="number" class="form-control" name="Sugar" required>
            </div>
            <div class="mb-3">
                <label>Cholesterol</label>
                <input type="number" class="form-control" name="Cholesterol" required>
            </div>
            <div class="mb-3">
                <label>Smoking</label>
                <select class="form-control" name="Smoking" required>
                    <option>Yes</option>
                    <option>No</option>
                </select>
            </div>
            <div class="mb-3">
                <label>Family History</label>
                <select class="form-control" name="FamilyHistory" required>
                    <option>Yes</option>
                    <option>No</option>
                </select>
            </div>
            <button type="submit" class="btn btn-primary w-100">Predict</button>
        </form>
    </div>
</div>
</body>
</html>
"""

result_template = """
<!DOCTYPE html>
<html>
<head>
    <title>Prediction Result</title>
    <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0/dist/css/bootstrap.min.css" rel="stylesheet">
</head>
<body class="bg-light">
<div class="container mt-5">
    <div class="card shadow-lg p-4 text-center">
        <h2 class="text-success">Prediction Result</h2>
        <h3 class="text-primary">{{ result }}</h3>
        <a href="/" class="btn btn-secondary mt-3">Back</a>
    </div>
</div>
</body>
</html>
"""
@app.route('/')
def home():
    return render_template_string(form_template)

@app.route('/predict', methods=['POST'])
def predict():
    try:

        age = int(request.form['Age'])
        gender = request.form['Gender'].strip().title()
        bmi = float(request.form['BMI'])
        bp = float(request.form['BP'])
        sugar = float(request.form['Sugar'])
        cholesterol = float(request.form['Cholesterol'])
        smoking = request.form['Smoking'].strip().title()
        family = request.form['FamilyHistory'].strip().title()


        user_data = [
            age,
            label_encoders['Gender'].transform([gender])[0],
            bmi,
            bp,
            sugar,
            cholesterol,
            label_encoders['Smoking'].transform([smoking])[0],
            label_encoders['FamilyHistory'].transform([family])[0]
        ]


        prediction = model.predict([user_data])
        predicted_label = label_encoders['Disease'].inverse_transform(prediction)[0]

        return render_template_string(result_template, result=predicted_label)

    except Exception as e:
        return f"<h2>Error: {str(e)}</h2><a href='/'>Back</a>"


if __name__ == '__main__':
    app.run(debug=True)
```
